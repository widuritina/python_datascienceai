```python
from randomuser import RandomUser
import pandas as pd

r=RandomUser()
user_list=r.generate_users(10)
for user in user_list:
    print(f"{user.get_full_name()},{user.get_email()},{user.get_picture()}")

def get_user():
    user_list=[]
    for user in RandomUser.generate_users(10):
        user_list.append({'Name':user.get_full_name(),'Email':user.get_email(),'Pict':user.get_picture()})
    return pd.DataFrame(user_list)
df1=get_user()
df1

```

```python
#Fruityvice API
import requests
import json

resp=requests.get("https://fruityvice.com/api/fruit/all")
result=json.loads(resp.content)
pd.DataFrame(result)

df=pd.json_normalize(result)
cherry=df.loc[df['name'] == 'Cherry']
print(f"Cherry family {cherry.iloc[0]['family']}, genus {cherry.iloc[0]['genus']}")
banana=df.loc[df['name']=='Banana']
print(f"Banana nutritions calorie {banana.iloc[0]['nutritions.calories']}")

resp=requests.get("https://official-joke-api.appspot.com/jokes/ten")
result=json.loads(resp.content)
pd.DataFrame(result)


```
