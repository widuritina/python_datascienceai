```python
#Data Loading
import pandas as pd

data=[10,20,30,40,50]
s=pd.Series(data)
print(s)
print(f"Accesing by index: {s[2]}")
print(f"Accessing by index position: {s.iloc[2]}")
print(f"Accessing multiple element:\n{s[1:3]}\n")


# Creating a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 28],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}
df=pd.DataFrame(data)
print(df)
#print(df['Name'])
print(df.iloc[2])
print(df.loc[2,'Name'])
print(df[['Name','City']]) #slicing dataframes coloumn
print(df[1:3])               #slicing rows

unique_age=df['Age'].unique()
print(unique_age)
df.to_csv('trading_data.csv')

```
